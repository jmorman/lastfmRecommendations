```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import json
import time
from operator import itemgetter
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df = pd.read_csv("lastfm-dataset-360K/usersha1-artmbid-artname-plays.tsv", sep='\t', nrows=5000000)
df.head()
```

```python
#List of top 2000 artists in first 100000 rows by artist_id
topfile = open("./outputs/top_2000_from_100000_prepared")
artists = [json.loads(line) for line in topfile][0]
```

```python
def jaccard_distance(aid1, aid2, m=50000, r=5):
    sample = df[:m]
    aid1_users = list(sample[sample.artist_id == aid1].user_id)
    aid2_users = list(sample[sample.artist_id == aid2].user_id)
    intersection = set(aid1_users) & set(aid2_users)
    union = set(aid1_users) | set(aid2_users)
    #Size of intersection over (size of union plus regularization)
    jaccard_sim = float(len(intersection))/(len(union)+r)
    #return 1-jaccard_sim #Standard Jaccard distance
    if(jaccard_sim == 0):
        return 10000 #generic large distance
    else:
        return 1/jaccard_sim - 1 #Maybe tease out more separation this way
```

```python
#Jaccard distance for Radiohead and The Beatles as a function of the number of rows of the dataset taken into consideration
dists = [jaccard_distance(artists[0][0], artists[1][0], m=k*20000) for k in range(1,251)]
```

```python
#Daft Punk and Depeche Mode
dists2 = [jaccard_distance(artists[10][0], artists[11][0], m=k*20000) for k in range(1,251)]
```

```python
#Beastie Boys and Nickelback
dists3 = [jaccard_distance(artists[100][0], artists[101][0], m=k*20000) for k in range(9,251)]
```

```python
#Patti Smith and Slowdive
#This indicates that for less popular artists the jaccard distance can still be fluctuating significantly at 5m rows
dists4 = [jaccard_distance(artists[1000][0], artists[1001][0], m=k*20000) for k in range(1,251)]
```

```python
plt.figure(figsize=(20,20))
plt.subplot(4,1,1)
plt.plot([np.nan]+dists)
plt.subplot(4,1,2)
plt.plot([np.nan]+dists2)
plt.subplot(4,1,3)
plt.plot([np.nan]*9+dists3)
plt.subplot(4,1,4)
plt.plot([np.nan]+dists4)
```

```python
def make_dists(n=10, show_time=True, rows=50000):
    pdists = []
    sample = artists[:n]
    start = time.time()
    for artist1 in sample:
        for artist2 in sample:
            if artist1 < artist2:
                pdists.append(jaccard_distance(artist1[0], artist2[0], m=rows))
    if(show_time):
        print "make_dists took " + str(time.time()-start) + " seconds"
    return pdists
```

```python
def plot_dendro(fig_size=(10,6), leaves=10, font=12., nolabels=False, m=50000):
    plt.figure(figsize=fig_size)
    dendrogram(
           linkage(make_dists(n=leaves, rows=m), 'average'),
           leaf_rotation=89.,
           leaf_font_size=font,
           labels=[artist[1][:15] for artist in artists][:leaves],
           distance_sort=True,
           no_labels=nolabels)
    plt.show()
```

```python
plot_dendro(leaves=10, m=50000)
```
