```python
pearson_sims_50k = [[json.loads(val) for val in line.strip().split("\t")] for line in open("./outputs/50k_MR_baked.out")]
pearson_sims_500k = [[json.loads(val) for val in line.strip().split("\t")] for line in open("./outputs/500k_MR_baked.out")]
pearson_sims_1m = [[json.loads(val) for val in line.strip().split("\t")] for line in open("./outputs/1m_MR_baked.out")]
```

```python
#not really a distance but it works
def convert_to_distance(pearson_sim):
    return 2.0/(pearson_sim+1)-1
```

```python
#List of top 2000 artists in first 100000 rows by artist_id
topfile = open("./outputs/top_2000_from_100000_prepared")
artists = [json.loads(line) for line in topfile][0]
```

```python
print list(artists)[:5]
```

```python
#Make pairwise distances for selected artists
#Execution time in sec:
#Rows\Artists
#      10   25   50   100  200  500  1000
#50k : 0.5  3    13   52   190  1101 4455
#500k: 6    39   152  603  2642
#1m  : 12   74   301  1363

def make_dists(n=10, sims=pearson_sims_50k, show_time=True):
    pdists = []
    sample = artists[:n]
    start = time.time()
    for artist1 in sample:
        for artist2 in sample:
            if artist1 < artist2:
                sim = [line for line in sims if ((line[0][0] == artist1[0]) and line[0][1] == artist2[0])]
                if(len(sim) < 1):
                    #Generic fill-in
                    dist = convert_to_distance(-0.9)
                else:
                    dist = convert_to_distance(sim[0][1])
                pdists.append(dist)
    if(show_time):
        print "make_dists took " + str(time.time()-start) + " seconds"
    return pdists
```

```python
pdists = make_dists()
artist_linkages = linkage(pdists, 'average')
print artist_linkages[:5]
print sorted(dist for dist in pdists if dist < 19)[-10:]
```

```python
def plot_dendro(fig_size=(10,6), leaves=10, simlist=pearson_sims_50k, font=12., nolabels=False):
    plt.figure(figsize=fig_size)
    dendrogram(
           linkage(make_dists(n=leaves, sims=simlist), 'average'),
           leaf_rotation=89.,
           leaf_font_size=font,
           labels=[artist[1][:15] for artist in artists][:leaves],
           distance_sort=True,
           no_labels=nolabels)
    plt.show()
```

```python
# generate two clusters: a with 100 points, b with 50:
np.random.seed(4711)  # for repeatability of this tutorial
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
X = np.concatenate((a, b),)
print X.shape  # 150 samples with 2 dimensions
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
print X[:10,:]
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(X, 'ward')
print type(Z)
```

```python
print Z[0] 
#Says: merged 52nd and 53rd clusters (points since this is beginning_, which had dist of 0.0415, 
#creating a cluster of 2 pts
```

```python
# calculate full dendrogram
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
plt.title('Hierarchical Clustering Dendrogram (truncated)')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=10,  # show only the last p merged clusters
    show_leaf_counts=True,  # numbers in brackets are counts
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,  # to get a distribution impression in truncated branches
)
plt.show()
```
